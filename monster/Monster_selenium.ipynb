```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
```

```python
# Find companies
def findCompanies(results):
    companies = results.findAll('div', class_ = "company")
    companyList = [(company.get_text().strip()) for company in companies]
    companyDf = pd.DataFrame({"Company":companyList})
    return companyDf
```

```python
# Find job titles
def findTitles(results):
    titles = results.findAll('h2', class_ = "title")
    titleList = [(title.get_text().strip()) for title in titles]
    titleDf = pd.DataFrame({"Title":titleList})
    return titleDf
```

```python
# Find job locations
def jobLocations(results):
    locations = results.findAll('div', class_ = "location")
    locationList = [(location.get_text().strip()) for location in locations]
    locationDf = pd.DataFrame({"Location":locationList})
    return locationDf
```

```python
# Find days
def findDays(results):
    days = results.findAll('time')
    dayList = [(day.get_text().strip()) for day in days]
    daysDf = pd.DataFrame({"Days":dayList})
    return daysDf
```

```python
# Extract contents from page
def extract_content(results):
    comp = findCompanies(results)
    title = findTitles(results)
    loc = jobLocations(results)
    day = findDays(results)
    jobDetails = pd.concat([comp,title,loc,day],axis = 1)
    return jobDetails
```

```python
def search_monster(jobSearchName, locationSearchName):
    time.sleep(10)
    jobName = driver.find_element_by_id("keywords2")
    jobName.send_keys(Keys.CONTROL,"a", Keys.DELETE);
    jobName.send_keys(jobSearchName)
    locationName = driver.find_element_by_id("location")
    locationName.send_keys(Keys.CONTROL,"a", Keys.DELETE);
    locationName.send_keys(locationSearchName)
    searchButton = driver.find_element_by_id("doQuickSearch")
    searchButton.click()
    time.sleep(10)
```

```python
def apply_filter():
    filterButton = driver.find_element_by_id("filter-flyout")
    filterButton.click()
    #driver.find_element_by_xpath("//*[@id='FilterCompany']/option[@value = 'cmpv=25720']").click()
    driver.find_element_by_xpath("//*[@id='FilterJobStatus']/option[@value = 'Full-Time']").click()
    time.sleep(10)
    filterSearchButton = driver.find_element_by_id("use-filter-btn")
    filterSearchButton.click()
    time.sleep(10)
```

```python
# Extracting contents through selenium for multiple form entries and applying filter
jobSearchNames = ["Finance", "Software"]
locationSearchNames = ["Austin", "Texas City"]
numberOfSearch = len(jobSearchNames)

#Initialising the driver
driver = webdriver.Chrome()
page = driver.get("https://www.monster.com/jobs/search")

for i in range(numberOfSearch):
    search_monster(jobSearchNames[i],locationSearchNames[i])
    apply_filter()
    # Extraction
    soup = BeautifulSoup(driver.page_source,'html.parser')
    results = soup.find(id = "ResultsContainer")
    contents = extract_content(results)
    # Printing
    print(contents)
    # Export to csv
    contents.to_csv('monster_job.csv', mode = 'a')
    time.sleep(10)
```
